```python
def read_json_string(path):
    '''
    Parameters
    -----------------
    path: str
        path = 'jawiki-country.json'
    articles2: list
        記事のリスト
    Return
    -----------------
    article
    '''
    article = []

    with open(path,'r') as articles:
        for i,a in enumerate(articles):
                article.append(json.loads(a))

    return article
```

```python
def get_one_text(article,text_number):
    '''
    get article from article num
    '''

    n = text_number

    text = article[n]['text']
    text = text.splitlines()
    
    return text
```

```python
def return_text_title(article):
    '''
    get all text　and title
    '''
    title = []
    text = []
    for i in range(len(article)):
        title.append(article[i]['title'])
        text.append(article[i]['text'])
        
    return title,text
```

```python
def get_info_list(text):
    '''
    get {{基礎情報}} from list
    '''
    info_list = []
    flag = 0
    pt_1 = r"\{\{\s?基礎情報"
    pt_0 = r'\'\'\'.{2,12}国\'\'\'（'
    check_country = []
    
    for tex in text:
        tex = tex.splitlines()
        for t in tex:
            if findall_pt(pt_1,t):
                flag = 1
            elif findall_pt(pt_0,t):
                flag = 0
                # 入ってない19個分の国を確認する
                m =  findall_pt(pt_0,t)
                check_country.append(m)
            elif flag==1:
                info_list.append(t)
# 一個のみtextを出力する場合は下記のコメントアウトを外す
#         else:
#             break
        
    return info_list,check_country
```

```python
def findall_pt(pattern,string):
    '''
    expect
    pt =  r"\{\{ 基礎情報 \}\}" or \{\{\s?基礎情報
    pt = r'\'\'\'.{3,12}国\'\'\'（'
    ?:findallである意味はあるのか?matchでもいいのでは
    '''
    prog = re.compile(pattern)
    m = re.findall(pattern,string)
    return m
# text一個ずつ入ってくる:エジプト→オーストラリア→インドネシアの順
# u = []
# 残り17こを確かめるために国名も格納する。
# 正規表現のパターン
    
```

```python
def chg_list_todict(temp_list:list)->dict:
    
    key_list = []
    value_list = []

    for i in temp_list:
        if  '=' in i:
            splited = i.split('=') 
            # splited = splited[0].strip()
            key_list.append(splited[0])
            value_list.append(splited[1])
        else:
            continue
    
    temp_dict = dict(zip(key_list,value_list))
    for key,value in temp_dict.items():
        print("{0}:{1}".format(key,value))
        
    return temp_dict
```

```python
def remove_markup(pattern,text):
    
        # 辞書型の場合

    if type(text) != 'list':
        renew = {}
        text = list(text.items())

        for k,v in text: 
            k = re.sub(pattern,"",k)
            v = re.sub(pattern,"",v)
            dictionary = {k:v}
            renew.update(dictionary)
        return renew
    else:
        # str型の場合
        renew = []

        for t in text:       
            t = re.sub(pattern,"",t)
            renew.append(t)
        
        return renew
```

```python
def remove_all_markup(text):
    # ''などのマークアップを削除する
    
    pattern26 =  r"\'\'\'(?!')|\'\'(?!')|\'(?!')"
    pattern = pattern26

    new26 = remove_markup(pattern,text)

    # [[|]]のパターン
    pattern27_2 = r"|\[\[.{10}\|(?!\[){2}.{10}\]\]|\[{2}|\]{2}"
    # |のパターン
    pattern27_3 = "|\||\[|\]" 
    pattern =   pattern27_2 +pattern27_3
    
    new27 = remove_markup(pattern,new26)

    # {{ }}はけす

#     pattern28_1 = r"\{\{\s?|\s?\}\}"
#     pattern28_2 = "|\</?ref\>?"
#     pattern28_3 = "|</?{[a-z]{2,12}\?/?>?|redirect"
#     pattern28_4 = "|<references />|redirect"
#     pattern28_2 = "|\</?ref\w{0,6}\/>| \/>|<\w{2}|redirect"
    pattern28_3= r"<([^>]*>|ref.+)|redirect"



    pattern = pattern28_3

    new28 = remove_markup(pattern,new27)


    return new28
```
